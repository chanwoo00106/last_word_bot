```python
import pandas as pd
import numpy as np
```

```python
url = 'https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx'
ranking = pd.read_html(url, encoding='utf-8')
pd_ranking = ranking[0].set_index('순위')
pd_ranking
```

```python
H_rank = pd_ranking.sort_values(by=['HR'], ascending=False)
H_rank
H_rank['안타 순위'] = np.arange(1, 31)
H_rank = H_rank.sort_values(by=['H'], ascending=False)
H_rank['홈런 순위'] = np.arange(1, 31)
H_rank[H_rank["HR"] >= 20]
```

```python
top30 = pd_ranking.head(30)[["선수명","AVG", "팀명", "H", "HR"]]
top30
```

```python
H_chart1 = top30.sort_values(by=['H'], ascending=False)
```

```python
H_chart1['안타 순위'] = np.arange(1, 31)
```
