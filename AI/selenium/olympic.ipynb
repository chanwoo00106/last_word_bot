```python
import pandas as pd

url = 'https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84'
```

```python
olympic = pd.read_html(url, encoding='utf-8', header=0, index_col=0)[0]
olympic
```

```python
summer=olympic[['하계 참가 횟수', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4']]
```

```python
summer.columns = ['참가수', '금', '은', '동']
summer
```

```python
summer['메달 총계'] = summer['금'] + summer['은'] + summer['동']
summer
```

```python
# 금, 은, 동 순으로 정렬
summer.sort_values(['금', '은', '동'], ascending=False)
```

```python
summer.to_excel('하계올림픽메달.xlsx')
```

```python
summer['금비중'] = summer['금'] / summer['메달 총계']

summer.sort_values('금비중', ascending=False)
```
