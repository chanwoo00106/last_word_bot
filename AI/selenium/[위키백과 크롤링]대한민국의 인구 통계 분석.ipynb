```python
import pandas as pd
```

```python
url = 'https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC'
tables = pd.read_html(url, encoding='utf-8') # cp949, euc-kr, utf-8
```

```python
poppulation = tables[5] # 인구통계 데이터프레임
poppulation
```

```python
#################################################
# 출생, 사망 분석
#################################################
import seaborn as sns
import matplotlib.pyplot as plt
# 한글 설정
sns.set(font="Malgun Gothic")
# sns.set(font="AppleGothic")
```

```python
poppulation["출생자수(명)"].plot()
```

```python
sns.pointplot(data=poppulation, x="연도 (년)", y='추계인구(명)')
```

```python
# 그래프 사이즈 조정
plt.figure(figsize=(15,4))
plt.xticks(rotation=60)
sns.pointplot(data=poppulation, x="연도 (년)", y='추계인구(명)')
```

```python
plt.figure(figsize=(30,8))
plt.xticks(rotation=60)
sns.pointplot(data=poppulation, x="연도 (년)", y='추계인구(명)')
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=60)
sns.lineplot(data=poppulation, x="연도 (년)", y='추계인구(명)')
```

```python
### 출생자 수
plt.figure(figsize=(12,3))
plt.xticks(rotation=60)
sns.pointplot(data=poppulation, x="연도 (년)", y='출생자수(명)')
```

```python
### 출생자 수
plt.figure(figsize=(12,3))
plt.xticks(rotation=60)
sns.lineplot(data=poppulation, x="연도 (년)", y='출생자수(명)')
```

```python
### 사망자 수
plt.figure(figsize=(12,3))
plt.xticks(rotation=60)
sns.lineplot(data=poppulation, x="연도 (년)", y='사망자수(명)')
```

```python
##############################################
# 인구 통계 추세 분석
##############################################
plt.figure(figsize=(15,8)) # 비율을 세로로 키움
plt.xticks(rotation=60)      # 글씨를 60도로 보일 수 있도록
sns.pointplot(data=poppulation, x='연도 (년)', y='출생자수(명)')
sns.pointplot(data=poppulation, x='연도 (년)', y='사망자수(명)')
```
